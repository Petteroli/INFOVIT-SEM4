```python

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.karate_club_graph()
```

```python
degrees = G.degree()
nodes = G.nodes()

print(G.nodes(data = True))
print(G.edges(data = True))
```

```python
colors = {
    "Mr. Hi" : "yellow",
    "Officer" : "blue"
}

n_color = [colors[nodes[n]['club']] for n in G.nodes]
n_size = np.asarray([degrees[n]*60 for n in nodes])


options = {
    "node_color": n_color,
    "node_size": n_size,
    "edge_color": "grey",
    "linewidths": 0,
    "width": 0.4,
    "with_labels" : True,
}
nx.draw_spring(G,**options)
```

```python
dijkstra_path = nx.dijkstra_path(G, 24, 16, weight='weight')
print(dijkstra_path)
```

```python
colors = {
    "Mr. Hi" : "yellow",
    "Officer" : "blue"
}

n_color = [colors[nodes[n]['club']] for n in G.nodes]
n_size = np.asarray([degrees[n]*60 for n in nodes])

for n in G.nodes:
    if n in dijkstra_path:
        n_color[n] = "red"

options = {
    "node_color": n_color,
    "node_size": n_size,
    "edge_color": "grey",
    "linewidths": 0,
    "width": 0.4,
    "with_labels" : True,
}
nx.draw_spring(G,**options)
```
